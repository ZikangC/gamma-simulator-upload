```python
#When you see this example, it means that you want to customize the energy spectrum. The most important parameter is source. 
#Replace the value after "hist energy" in the code below with the value corresponding to the X-axis scale of your customized energy spectrum
from gamma_simulator import gamma_simulator

simulator = gamma_simulator(verbose=True,
                            verbose_plots={'energy': True, 'signal': True},
                            source={'hist_energy':[1,2,3,4,5],
                                    'hist_counts':[3,6,8,4,6]},
                            signal_len=1, 
                            fs=1e6,
                            lambda_value=1000,
                            dict_type='gamma',
                            dict_shape_params={'mean1': 0.1,
                                              'std1': 0.001,
                                              'mean2': 1e5,
                                               'std2': 1e3},
                            noise_unit='std',
                            noise=1e-5,
                            dict_size=10,
                            seed=42)
s = simulator.generate_signal()
```
